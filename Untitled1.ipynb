```python
"""
Dumps folder structure + all file contents into a single text file.
Usage: python dump_project.py
"""
import os
from pathlib import Path

OUTPUT_FILE = "project_dump.txt"

EXCLUDE_DIRS = {'venv', '.git', '__pycache__', 'mlruns', '.ipynb_checkpoints',
                'node_modules', '.pytest_cache', 'screenshots', 'models'}
EXCLUDE_EXTS = {'.pyc', '.joblib', '.png', '.jpg', '.jpeg', '.gif', '.csv',
                '.pkl', '.zip', '.exe', '.dll'}
INCLUDE_EXTS = {'.py', '.yml', '.yaml', '.md', '.txt', '.json', '.toml',
                '.cfg', '.ini', '.ipynb', '.sh', '.html'}
INCLUDE_NAMES = {'Dockerfile', '.gitignore', '.dockerignore', 'Makefile'}

def build_tree(root, prefix=""):
    """Generate an ASCII folder tree."""
    lines = []
    entries = sorted(
        [e for e in root.iterdir() if e.name not in EXCLUDE_DIRS],
        key=lambda x: (x.is_file(), x.name.lower())
    )
    for i, entry in enumerate(entries):
        connector = "└── " if i == len(entries) - 1 else "├── "
        lines.append(f"{prefix}{connector}{entry.name}")
        if entry.is_dir():
            extension = "    " if i == len(entries) - 1 else "│   "
            lines.extend(build_tree(entry, prefix + extension))
    return lines

def should_include(path: Path) -> bool:
    if any(part in EXCLUDE_DIRS for part in path.parts):
        return False
    if path.suffix.lower() in EXCLUDE_EXTS:
        return False
    if path.name in INCLUDE_NAMES:
        return True
    return path.suffix.lower() in INCLUDE_EXTS

def main():
    root = Path.cwd()
    output = []

    # Header
    output.append("=" * 70)
    output.append(f"PROJECT DUMP: {root.name}")
    output.append(f"Root: {root}")
    output.append("=" * 70)

    # Folder structure
    output.append("\n" + "=" * 70)
    output.append("FOLDER STRUCTURE")
    output.append("=" * 70 + "\n")
    output.append(root.name + "/")
    output.extend(build_tree(root))

    # File contents
    output.append("\n\n" + "=" * 70)
    output.append("FILE CONTENTS")
    output.append("=" * 70)

    files = sorted([p for p in root.rglob("*") if p.is_file() and should_include(p)])

    for file in files:
        rel_path = file.relative_to(root)
        output.append("\n" + "─" * 70)
        output.append(f"📄 FILE: {rel_path}")
        output.append("─" * 70)
        try:
            content = file.read_text(encoding="utf-8", errors="replace")
            output.append(content)
        except Exception as e:
            output.append(f"[Could not read file: {e}]")

    # Write
    Path(OUTPUT_FILE).write_text("\n".join(output), encoding="utf-8")
    print(f"✅ Dumped {len(files)} files into {OUTPUT_FILE}")
    print(f"   Size: {Path(OUTPUT_FILE).stat().st_size / 1024:.1f} KB")

if __name__ == "__main__":
    main()
```
